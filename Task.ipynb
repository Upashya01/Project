#***Analysis***

# **i) Batsman**

```python
plt.figure(figsize=(20,6))
runs=deliveries.groupby('batsman')['batsman_runs'].agg('sum').reset_index().sort_values(by='batsman_runs',ascending=False).reset_index(drop=True)
plt.title("Highest Runs till now")
sns.barplot(x=runs["batsman"].iloc[:10],y=runs["batsman_runs"].iloc[:25],data=runs)
plt.xlabel("Batsman")
plt.ylabel("Runs")
plt.show()
```

```python
plt.figure(figsize=(20,6))
Fours=deliveries.groupby('batsman')['batsman_runs'].agg(lambda x:(x==4).sum()).reset_index().sort_values(by='batsman_runs',ascending=False).reset_index(drop=True)
plt.title("Fours")
sns.barplot(x=Fours["batsman"].iloc[:10],y=Fours["batsman_runs"].iloc[:20],data=Fours)
plt.xlabel("Batsman")
plt.ylabel("Total Number of Fours")
plt.show()
```

```python
plt.figure(figsize=(20,6))
Sixes=deliveries.groupby('batsman')['batsman_runs'].agg(lambda x:(x==6).sum()).reset_index().sort_values(by='batsman_runs',ascending=False).reset_index(drop=True)
plt.title("Sixes")
sns.barplot(x=runs["batsman"].iloc[:10],y=runs["batsman_runs"].iloc[:20],data=runs)
plt.xlabel("Batsman")
plt.ylabel("Total Number of Sixes")
plt.show()
```

```python
least_runs=runs.tail(25)
plt.figure(figsize=(16,6))
plt.title("Least Runs till now")
sns.barplot(x=least_runs["batsman"].iloc[:10],y=least_runs["batsman_runs"].iloc[:10],data=least_runs)
plt.xlabel("Batsman")
plt.ylabel("Runs")
plt.show()
```

# ii) **Bowler**


```python
plt.figure(figsize=(20,6))
balls=deliveries.groupby('bowler')['ball'].agg('count').reset_index().sort_values(by='ball',ascending=False).reset_index(drop=True)
plt.title("Highest balls till now")
sns.barplot(x=balls["bowler"].iloc[:20],y=balls["ball"].iloc[:25],data=balls)
plt.xlabel("Bolwer")
plt.ylabel("Total Balls")
plt.show()
```

```python
plt.figure(figsize=(20,6))
dot_balls=deliveries.groupby('bowler')['total_runs'].agg('count').reset_index().sort_values(by='total_runs',ascending=False).reset_index(drop=True)
plt.title("Most Dot Balls")
sns.barplot(x=dot_balls["bowler"].iloc[:15],y=dot_balls["total_runs"].iloc[:25],data=dot_balls)
plt.xlabel("Bolwer")
plt.ylabel("Total Dot Balls")
plt.show()
```

```python
plt.figure(figsize=(20,6))
extras=deliveries.groupby('bowler')['extra_runs'].agg(lambda y:(y>0).sum()).reset_index().sort_values(by='extra_runs',ascending=False).reset_index(drop=True)
plt.title("Highest Extra balls till now")
sns.barplot(x=extras["bowler"].iloc[:15],y=extras["extra_runs"].iloc[:20],data=extras)
plt.xlabel("Bolwer")
plt.ylabel("Extras Balls")
plt.show()
```

#**iii) Team**

```python
team_m=pd.melt(matches,id_vars=['id','season'],value_vars=['team1','team2'])
plt.figure(figsize=(22,8))
sns.countplot(x='value',data=team_m,order=team_m['value'].value_counts().index)
plt.title("Highest number of matches played by a team")
plt.xlabel("Teams")
plt.ylabel("Matches")
plt.xticks(rotation=90)
plt.show()
```

```python

plt.figure(figsize=(22,8))
sns.countplot(x='winner',data=matches,order=matches['winner'].value_counts().index)
plt.title("Highest wins by a team")
plt.xlabel("Teams")
plt.ylabel("Total Wins")
plt.xticks(rotation=90)
plt.show()
```
